# 1. Mezcla de fertilizantes

```python
print("\n Variables de decision:")
print("x1 = toneladas de fertilizante F1 a producir")
print("x2 = toneladas de fertilizante F2 a producir")

print("\n Funcion objetivo:")
print("Max Z = 180*x1 + 220*x2")

print("\n Restricciones:")

print("Compuesto A: 3*x1 + 4*x2 <= 120")
print("Compuesto B: 2*x1 + 1*x2 <= 80")
print("Compuesto C: 5*x1 + 3*x2 <= 150")

print("\nNo negatividad:")
print("x1 >= 0")
print("x2 >= 0")

print("\n Clasificacion:")
print("Modelo lineal: todas las ecuaciones son de primer grado")
print("Modelo continuo: las variables pueden tomar valores reales")
```

```python
import numpy as np
from scipy.optimize import linprog

c = [-180, -220]

A = [
    [3, 4],   # Compuesto A
    [2, 1],   # Compuesto B
    [5, 3]    # Compuesto C
]

b = [120, 80, 150]

bounds = [(0, None), (0, None)]

resultado = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

x1, x2 = resultado.x
Z = -resultado.fun  # volver a positivo

print("\nEstado:", "Optimo encontrado" if resultado.success else "Fallo")

print("\nSolucion optima:")
print(f"x1 (Fertilizante F1) = {x1:.2f} toneladas")
print(f"x2 (Fertilizante F2) = {x2:.2f} toneladas")

print(f"\nGanancia maxima = ${Z:.2f}")
```

# 2. Producción semanal de una fábrica

```python
S0 = 200
M0 = 80

print("\n Variables:")
print("Variables dependientes: S_n (sillas), M_n (mesas)")
print("Variable independiente: n (tiempo en semanas)")

print("\n Clasificacion:")
print("Modelo lineal: las ecuaciones son combinaciones lineales de S_n y M_n")
print("Modelo discreto: se define por semanas (n, n+1)")

S1 = 0.6*S0 + 0.2*M0 + 40
M1 = 0.1*S0 + 0.5*M0 + 20
S2 = 0.6*S1 + 0.2*M1 + 40
M2 = 0.1*S1 + 0.5*M1 + 20

print("\n Resultados:")
print(f"S1 = {S1:.2f}")
print(f"M1 = {M1:.2f}")
print(f"S2 = {S2:.2f}")
print(f"M2 = {M2:.2f}")

print("\n Sistema en equilibrio:")

print("S = 0.6*S + 0.2*M + 40")
print("M = 0.1*S + 0.5*M + 20")
```

# 3. Epidemia en una ciudad

```python
beta = 0.00003
gamma = 0.1
N = 10000

print("\n Variables y parametros:")
print("Variable: I(t) = numero de infectados")
print(f"beta = {beta}")
print(f"gamma = {gamma}")
print(f"Poblacion total = {N}")

print("\n El modelo es no lineal debido al termino (10000 - I)*I")

I0 = 50
print("\n Condicion inicial:")
print(f"I(0) = {I0}")

I1 = 0
I2 = N - gamma / beta

print("\n Puntos de equilibrio:")
print(f"I1 = {I1}")
print(f"I2 = {I2:.2f}")

# e) Interpretación
print("\n Interpretacion:")
print("I = 0 -> No hay enfermedad")
print("I ≈ 6667 -> Enfermedad se mantiene (estado endemico)")
```

# 4. Modelo logístico de pesca

```python
import math

r = 0.4
K = 1000
H = 80
P0 = 400

print("\n Variables y parametros:")
print("Variable: P_n = poblacion en el año n")
print(f"r = {r}")
print(f"K = {K}")
print(f"H = {H}")

print("\n El modelo es no lineal (termino P*(1 - P/K)) y discreto")

P1 = P0 + r * P0 * (1 - P0 / K) - H
P2 = P1 + r * P1 * (1 - P1 / K) - H

print("\n Resultados:")
print(f"P1 = {P1:.2f}")
print(f"P2 = {P2:.2f}")

a = 1
b = -1000
c = 200000

disc = b**2 - 4*a*c

P_eq1 = (-b + math.sqrt(disc)) / (2*a)
P_eq2 = (-b - math.sqrt(disc)) / (2*a)

print("\n Puntos de equilibrio:")
print(f"P1 = {P_eq1:.2f}")
print(f"P2 = {P_eq2:.2f}")

# e) Interpretación
print("\n Interpretacion:")
print("P = 276 -> equilibrio inestable (riesgo de extincion)")
print("P = 724 -> equilibrio estable (poblacion sostenible)")
```
